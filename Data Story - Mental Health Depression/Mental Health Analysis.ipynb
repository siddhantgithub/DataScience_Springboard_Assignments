## Part 1 - Data Wrangling

### Read mental health data

```python
mh_df = pd.read_csv("Mental health Depression disorder Data.csv")
```

```python
mh_df.head(5)
```

```python
mh_df.info()
```

```python
mh_df.dropna(inplace=True)
```

```python
mh_df.isnull().sum()
```

```python
mh_df.Year.value_counts()
```

```python
mh_df.Year = pd.to_numeric(mh_df.Year, errors='coerce')
```

```python
mh_df.Year.describe()
```

### We only select data for years between 1990 and 2017 (both including) as they have complete information

```python
mh_df = mh_df[mh_df.Year >= 1990 and mh_df.Year < 2018]
```

```python
mh_df.Year.value_counts()[mh_df.Year.value_counts() == 1012]
```

### Convert all numbers to floats

```python
for x in ['Schizophrenia (%)', 'Bipolar disorder (%)', 'Eating disorders (%)']:
    mh_df[x] =  mh_df[x].astype(float)
```

```python
mh_df.describe()
```

### Calculating correlation matrix among different mental health conditions

```python
mh_df[['Schizophrenia (%)',
       'Bipolar disorder (%)', 'Eating disorders (%)', 'Anxiety disorders (%)',
       'Drug use disorders (%)', 'Depression (%)', 'Alcohol use disorders (%)']].corr()
```

### Calculating total percentage and pivoting to get the data in the right shape

```python
mh_df['Total_Percentage'] = mh_df[['Schizophrenia (%)',
       'Bipolar disorder (%)', 'Eating disorders (%)', 'Anxiety disorders (%)',
       'Drug use disorders (%)', 'Depression (%)',
       'Alcohol use disorders (%)']].sum(axis = 1)
```

```python
mh_df['Total_Percentage'].describe()
```

```python
mh_df_pivot = pd.pivot_table (mh_df, index = ['Entity','Code'], columns='Year', values = 'Total_Percentage').reset_index()
```

```python
mh_df_pivot['TPC'] = mh_df_pivot[2017] - mh_df_pivot[1990]
```

### Reading world GDP data and merging it with world health data using Country code

```python
gdp_df = pd.read_csv('GDP.csv')
```

```python
gdp_df.head()
```

```python
gdp_df_selected = gdp_df[[str(x) for x in range(1990,2018)]]
```

```python
gdp_df_selected['Country Code'] = gdp_df['Country Code']
```

```python
gdp_df_selected.head()
```

```python
mh_gdp_merge = mh_df_pivot.merge(gdp_df_selected, left_on = "Code", right_on = "Country Code", suffixes=('_mh', '_gdp')).dropna()
```

```python
mh_gdp_merge.head()
```

```python
mh_gdp_corr_df = mh_gdp_merge[['2017', 2017, 'TPC']]
mh_gdp_corr_df.loc[:,"Change in GDP in USD from 1990 to 2017"] = mh_gdp_merge['2017'] - mh_gdp_merge['1990']
mh_gdp_corr_df.columns = ['GDP in USD in 2017', '% of Population having mental health problem', '% change in population having mental health problems','Change in GDP (USD) from 1990 to 2017']
```
